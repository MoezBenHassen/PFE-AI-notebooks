```python
PredDataframe = pd.DataFrame([], columns=['Instrument','Date','Prediction'])
FEATURES =['RSI' ,'interRSI' ,'BollingerBands.middle' ,'BollingerBands.upper' ,'BollingerBands.lower' ,'BollingerBands.pb','interBollingerBands' ,
'ROC' ,'interROC' ,'MACD.MACD' ,'MACD.signal' ,'MACD.histogram' ,'interMACD' ,'MOM.MOM' ,'MOM.EmaMOM','interMOM' ,'ATR' ,'WEMA' ,'KST.kst' ,'KST.signal' ,'Stochastic.k' ,'Stochastic.d' ,
'WilliamsR' ,'OBV' ,'ADL' ,'TRIX' ,'chaikinVol','price'  ]                     
def getTaForInstru (instru, dDate, df=df):
   # print(df.dtypes)
    lastTradingData = df[(df['instrument'] == instru) & (df['date'] < dDate)].reset_index()
    return lastTradingData.loc[0]
LastTradingData = getTaForInstru('ADWYA', dDate=datetime(2021,12,24))


def extract(record ):
    feats = FEATURES
    filter = np.zeros(len(feats))
    X = []
    i = -1
    for feat in feats:
        i += 1
        if(filter[i]) :
            continue

        prop = feat.split('.')
        rec = record
        for p in prop:
            if isinstance(rec, dict):
                if p in rec.keys():
                    rec = rec[p] if(rec[p]) else 0
                else :
                    rec = 0
            else :
                rec = 0
        temp = np.float64(rec)
        if(not np.isfinite(temp)):
            temp = 0
        X.append(temp)
    return X

x = {}
for index, item in LastTradingData.items():
    if index not in ['index', '_id']:
        x[index] = item

X = [extract(x)]

X= np.array(X, np.float64)

prediction = CLASSIFIER.predict(X) 
print(int(prediction))
```
